## Design FIR Filter

Calculated filter coefficients are exported as a csv-file.

## Import Packages

```python
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt
```

## Define Parameters

```python
Fsmp = 1000             # Sampling Frequency
Fnyq = Fsmp / 2.0       # Nyquist Freqency
Fcut = 100.0 / Fnyq     # Cutoff Frequency (Normalized)
numtaps = 45            # Number of Filter Coefs
```

## Define & Check Filter Coefs

```python
b = signal.firwin(numtaps, Fcut)

b_reshaped = np.reshape(b, (9, 5))
np.savetxt('coef.csv', b_reshaped, fmt='% 7f', delimiter=',')
```

```python
plt.plot(b)
plt.show()
```

## Check Filter Response

```python
w, h = signal.freqz(b)
```

```python
plt.figure(figsize=(12, 5))  # (width, height)
plt.title('Digital filter frequency response')
plt.xscale("log")
 
a = 20 * np.log10(abs(h));
f = w / (2 * np.pi) * Fsmp
plt.plot(f, a  )
plt.title('Digital filter frequency response')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude [dB]')
plt.margins(0, 0.1)
plt.xlim(10, Fsmp)
plt.ylim([-15,15])
plt.grid(which='both', axis='both')
plt.show()
```
